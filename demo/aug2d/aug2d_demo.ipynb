## 2D Augmentor

It performs augmentation on 2D images. It is basically implemented to   perform on-the-fly augmentation. So, to implement a custom ImageDataGenerator or any data loader, it will come in handy. It performs following augmentation: <br><br>
        * Rotation * Flip * Zoom (original image size remains same) 
        * Shift * Rescale * Shear

```python
from aug2d import Augmentor2d
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Read image

```python
img = Image.open('/content/drive/MyDrive/Colab Notebooks/GitHub/dataLoader/dataset/images/0011.png')
img = np.array(img)
```

### Create an augmentation instance

```python
aug = Augmentor2d(img)
```

### Rotate image

```python
rot_img = aug.rotate(-45, mode='constant')

# Plot
fig, axs = plt.subplots(1,2)
axs[0].imshow(img)
axs[1].imshow(rot_img)
```

### Flip image.0 means horizontal, and 1 means vertical flip

```python
flip_img = aug.flip(flip_axis=1)

# Plot
fig, axs = plt.subplots(1,2)
axs[0].imshow(img)
axs[1].imshow(flip_img)
```

### Zoom image. For zoom-in, factor>1, for zoom-out, factor<1

```python
zoom_img = aug.zoom(factor=0.5) # zoom-out

# Plot
fig, axs = plt.subplots(1,2)
axs[0].imshow(img)
axs[1].imshow(zoom_img)
```

```python
zoom_img = aug.zoom(factor=1.8) # zoom-in

# Plot
fig, axs = plt.subplots(1,2)
axs[0].imshow(img)
axs[1].imshow(zoom_img)
```

### Shift image

```python
shift_img = aug.shift((55, 25), mode='constant')

# Plot
fig, axs = plt.subplots(1,2)
axs[0].imshow(img)
axs[1].imshow(shift_img)
```

### Shear image

```python
shear_img = aug.shear(factor=0.2)

# Plot
fig, axs = plt.subplots(1,2)
axs[0].imshow(img)
axs[1].imshow(shear_img)
```

### Rescale image

```python
rescale = aug.rescale(1/255.)
```
